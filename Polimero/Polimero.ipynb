```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
from pylab import *
import matplotlib.pyplot as plt
Data=pd.read_csv('Data/Ree_T=1_KOnM=1000_N=9.csv')

def f(F,N,b,T):
    return N*b*(1/np.tanh(F*b/T)-T/F/b)

#expected=(3,1,0.1)
#popt,cov=curve_fit(f,xDataRaw,yDataRaw,expected)

plt.figure(figsize=(10,10))         
xfine=np.linspace(-6,6,500)
plot(xfine,f(xfine,8,1,1),label='model',lw='1',color='blue')

plt.ylabel('Ree',fontsize='20')
plt.xlabel('Fuerza',fontsize='20')
plt.scatter(Data.loc[:,"F"],Data.loc[:,"Ree"],s=5,label='data',color='red')

plt.grid()
plt.legend(prop={'size':'20'})
plt.savefig("8Enlaces_KOnM=1000.png",format='png',ndpi=12000)
```
